# Spearman rank full vs one of extremes

```python
# the indices are added together into a st, we get the top and least 10 of each? Maybe do two correlations? ill try that one first at least

def getExtremeIDX(lst, largest=True, num=10):
    if largest:
        return list(reversed(sorted(range(len(lst)), key=lambda i: lst[i])[-num:]))
    else:
        return sorted(range(len(lst)), key=lambda i: lst[i])[:num]




    
    
```

```python
full_file = 'full-hpv'
one_file = 'one-hpv'
#one_file = 'all-train-tenth-hpv'
num=10
largest=True
```

```python
def extremeCor(full_file, one_file, num, largest):
    cors = []
    tops = []

    for i in range(10):
        full_p = getPreds(full_file, i)
        smol_p = getPreds(one_file, i)
        if len(full_p) != len(smol_p):
            continue

        top = getExtremeIDX(full_p, num=num, largest=largest) + getExtremeIDX(smol_p, num=num, largest=largest)

        top_full_p = [full_p[i] for i in top]
        top_smol_p = [smol_p[i] for i in top]

        tops.append((top_full_p, top_smol_p))

        c, p = stats.spearmanr(top_smol_p, top_full_p)
        cors.append(c)
    
    return cors
```

```python
# one, full
full_file = 'full-hpv'
one_file = 'one-hpv'
#one_file = 'all-train-tenth-hpv'
num=10
largest=True

cors = extremeCor(full_file, one_file, num, largest)

size = 10
plt.figure(figsize=(size, size / 2))
sns.histplot(cors)
plt.title("Correlation Div 1, no subset", size=size*3)
plt.xlabel("Spearman Rank Correlation of 10 largest influences", size=size*2)

plt.show()
plt.clf()
```

```python
# one, all train tenth
# one, full
full_file = 'full-hpv'
one_file = 'all-train-tenth-hpv'
#one_file = ''
num=10
largest=True

cors = extremeCor(full_file, one_file, num, largest)

size = 10
plt.figure(figsize=(size, size / 2))
sns.histplot(cors)
plt.title("Correlation div 10, no subset", size=size*3)
plt.xlabel("Spearman Rank Correlation of 10 largest influences", size=size*2)

plt.show()
plt.clf()
```

```python
# full, all train tenth
# one, full
full_file = 'one-hpv'
one_file = 'all-train-tenth-hpv'
#one_file = ''
num=10
largest=True

cors = extremeCor(full_file, one_file, num, largest)

size = 10
plt.figure(figsize=(size, size / 2))
sns.histplot(cors)
plt.title("Correlation div 10, div 1", size=size*3)
plt.xlabel("Spearman Rank Correlation of 10 largest influences", size=size*2)

plt.show()
plt.clf()
```

# Graph raw inf functions

```python

def getRetrainingLosses(file, idx):
    path = "../output/mnist/%s/mnist_small_all_cnn_c_loss_diffs-%s.npz" % (file, idx)
    if not os.path.exists(path):
        return False, False
        
    data = np.load(path, allow_pickle=True)
    lst = data.files
    
    stuff = {}

    for item in lst:
        stuff[item] = list(data[item])
    
    actual = stuff['actual_loss_diffs']
    predicted = stuff['predicted_loss_diffs']
        
    return predicted, actual
```

```python
def plotExtreme(root_file, comp_file, idx_range=10, retraining=False):
    data = []
    idxs = []
    for i in range(idx_range):
        if retraining:
            # Make actual the root and predicted the comparison
            comp_p, root_p = getRetrainingLosses(root_file, i)
        else:
            root_p = getPreds(root_file, i)
            comp_p = getPreds(comp_file, i)
        
        if len(root_p) != len(comp_p):
            continue
            
        idxs.append(i)
        top = getExtremeIDX(root_p, num=num, largest=largest)
        top_root_p = [root_p[i] for i in top]
        top_comp_p = [comp_p[i] for i in top]

        
        for n in range(len(top_root_p)):
            lab = '%s-%s' % (root_file, (n + 1))
            data.append({
                'test_idx': i,
                'label': lab,
                'root-inf': top_root_p[n],
                'comp-inf': top_comp_p[n]
            })
                 
    df = pd.DataFrame(data)
    
    
    return df, idxs
```

```python
comp_file = 'full-hpv'
root_file = 'one-hpv'
#one_file = 'all-train-tenth-hpv'
num=10
largest=True

root_file='retraining-tenth-hpv'
#root_file='retraining'#-tenth-hpv'

df, idxs = plotExtreme(root_file, comp_file, retraining=True)
```

```python
for idx in idxs:
    size = 10
    plt.figure(figsize=(size, size))
    num_df = df[df['test_idx'] == idx]
    plt.subplot(2, 1, 1)
    sns.barplot(data=num_df, x='label', y='root-inf', color='teal')
    
    plt.subplot(2, 1, 2)
    sns.barplot(data=num_df, x='label', y='comp-inf', color='teal')
    
    
    plt.show()
    plt.clf()
```
